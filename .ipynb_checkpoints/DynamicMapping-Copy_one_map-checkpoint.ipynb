```python
# import libraries
import pandas as pd
import geopandas as gpd
import PIL
import io

pd.set_option('display.max_rows', 250)
```

```python
# datenabruf online
df = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv")
```

```python
# df.head(1)
```

```python
# grouping data by country
df = df.groupby('Country/Region').sum()
df.head(1)
```

```python
# drop lat and long
df = df.drop(columns = ['Lat', 'Long'])
df.head(1)
```

```python
# create a tanspose of the df
df_T = df.T
df_T.head(3)
# df_transpond
```

```python
df_T.plot(y = ['Germany', 'China', 'Italy', 'US'],
          use_index = True, 
          figsize = (4,4),
         marker = '')
```

```python
# read worldmap shapefile
world = gpd.read_file('D:\Github\Covid-19-JHU-mapchart\Maps\World_Map.shp')
world
```

```python
# replace world shapefile with names from JHU
# replace (... with ...)
world.replace('Viet Nam', 'Vietnam', inplace = True)
world.replace('United States', 'US', inplace = True)
world.replace('Brunei Darussalam', 'Brunei', inplace = True)
world.replace('Cape Verde', 'Cabo Verde', inplace = True)
world.replace('Congo', 'Congo (Brazzaville)', inplace = True)
world.replace('Democratic Republic of the Congo', 'Congo (Kinshasa)', inplace = True)
world.replace('Czech Republic', 'Czechia', inplace = True)
world.replace('Swaziland', 'Eswatini', inplace = True)
world.replace('Holy See (Vatican City)', 'Holy See', inplace = True)
world.replace('Iran (Islamic Republic of)', 'Iran', inplace = True)
world.replace("Korea, Democratic People's Republic of", 'Korea, South', inplace = True)
world.replace("Lao People's Democratic Republic", 'Laos', inplace = True)
world.replace('Libyan Arab Jamahiriya', 'Libya', inplace = True)
world.replace('Republic of Moldova', 'Moldova', inplace = True)
world.replace('The former Yugoslav Republic of Macedonia', 'North Macedonia', inplace = True)
world.replace('Syrian Arab Republic', 'Syria', inplace = True)
world.replace('Taiwan', 'Taiwan*', inplace = True)
world.replace('United Republic of Tanzania', 'Tanzania', inplace = True)
world.replace('Palestine', 'West Bank and Gaza', inplace = True)              
world.replace('', '', inplace = True)              
             
```

```python
# check if shapefile and JHU data countrynames are the same
for index, row in df.iterrows():
    #print(index)
    if index not in world['NAME'].to_list():
        print(index + ' from JHU data is not in the list of countries of the shapefile')
    else:
        pass
```

```python
#  merge the 'df' (JHU) with'world' geopandas dataframe

merge = world.join(df, on = 'NAME', how='right')
merge
```

```python
#### plot

date = '1/4/21'

ax = merge.plot(column = date,
                cmap = 'YlOrRd',
                figsize = (20,20),
                legend=True,
                scheme = 'user_defined',
                classification_kwds = {'bins':[100,1000,10000,100000,1000000,10000000,100000000]},
                edgecolor = 'black',
                linewidth = 0.4)
# add title
ax.set_title('Total Confirmed Coronavirus Cases ' + date , fontdict = {'fontsize':20}, pad = 12.5)

leg = ax.get_legend()

for lbl in leg.get_texts():
    label_text = lbl.get_text()
    lower = label_text.split()[0]
    upper = label_text.split()[2]
    new_text = f'{float(lower):,.0f} - {float(upper):,.0f}'
    lbl.set_text(new_text)


# removing axes
ax.set_axis_off()

# move legend
ax.get_legend().set_bbox_to_anchor((0.16, 0.5))
```
